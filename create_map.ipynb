```python
import numpy as np
import pandas as pd
import geopandas as gpd
import folium
```

```python
df_drk = pd.read_excel("./Kleidersammlung.xlsx")
```

```python
# get coordinates of adresses
df_drk["address_lookup"] = df_drk.Adresse + " " + df_drk.Ort
df_geocode = gpd.tools.geocode(df_drk.address_lookup,  provider='nominatim', user_agent="red5alex_drk")

print(f"{df_drk.address_found.isna().sum()} Adressen nicht gefunden:\n")
print("\n".join(df_drk[df_drk.address_found.isna()].Adresse))

df_drk["address_found"] = df_geocode.address
df_drk["geometry"] = df_geocode.geometry
```

```python
df_drk[df_drk.address_found.isna()]
```

```python
# convert to GeoDataFrame
gdf_drk = gpd.GeoDataFrame(df_drk)
```

```python
# get starting location
center = gdf_drk.dissolve().centroid.loc[0]

# helper function to create the popup object
def popup_template(row):
    html = f"{row.Adresse}, {row.Ort}"
    if row.Bemerkung is not np.nan:
        html += f"<br>{row.Bemerkung}"
    popup = folium.map.Popup(html=html, max_width=200)
    return popup

# create map object
m = folium.Map(location=(center.y, center.x), 
               zoom_start=11,
              control_scale=True, zoom_control=True
              )

# add a marker for each entry with a valid Adress
for i, row in gdf_drk.dropna(subset="address_found").iterrows():
    m.add_child(folium.map.Marker(
        location = (row.geometry.y, row.geometry.x),
        popup = popup_template(row),
    ))

# save to disk
m.save("drk_map.html")
```
